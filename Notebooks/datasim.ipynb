## Orders Table

```python
orders = pd.read_csv("D:/E-commerce/Data/raw/olist_orders_dataset.csv")
orders_2=pd.read_csv("D:/E-commerce/Data/raw/olist_order_items_dataset.csv")
```

```python
fact = orders_2.merge(
    orders[['order_id','customer_id','order_purchase_timestamp']],
    on='order_id',
    how='left'
)
```

```python
fact.shape
```

```python
payments_agg = paymentrw.groupby('order_id').agg(
    payment_type=('payment_type','first'),  # or choose main payment type
    payment_value=('payment_value','sum')
).reset_index()

fact = fact.merge(payments_agg, on='order_id', how='left')
```

```python
payment_dim = payment[['payment_type']].drop_duplicates().reset_index(drop=True)
payment_dim['payment_key'] = payment_dim.index + 1

fact['payment_key'] = fact['payment_type'].map(
    dict(zip(payment_dim['payment_type'], payment_dim['payment_key']))
)
```

```python
# Merge product info first
fact = fact.merge(
    products[['product_id','product_category']],
    on='product_id',
    how='left'
)

# Now map the category key
fact['category_key'] = fact['product_category'].map(
    dict(zip(category['product_category'], category['category_key']))
)
```

```python
# Category
fact['category_key'] = fact['product_category'].map(
    dict(zip(category['product_category'], category['category_key']))
)

# Product
fact['product_key'] = fact['product_id'].map(
    dict(zip(products['product_id'], products['product_key']))
)

# Customer
fact['customer_key'] = fact['customer_id'].map(
    dict(zip(customers['customer_id'], customers['customer_key']))
)

# Shipping
fact['shipping_key'] = fact['order_id'].map(
    dict(zip(shipping['order_id'], shipping['shipping_key']))
)
```

```python
fact['order_purchase_timestamp'] = pd.to_datetime(fact['order_purchase_timestamp'])
fact['order_date'] = fact['order_purchase_timestamp'].dt.normalize()
fact['order_purchase_timestamp']=fact['order_purchase_timestamp'].dt.strftime('%Y-%m-%d' )

date_map = dict(zip(date['full_date'], date['date_key']))
fact['date_key'] = fact['order_date'].map(date_map)

fact = fact.drop(columns=['order_date'])
```

```python
fact['order_purchase_timestamp'].info()
```

```python
fact.shape
```

```python
fact
```

```python
fact['quantity'] = fact.groupby(['order_id','product_id'])['order_item_id'].transform('count')
```

```python
fact['quantity'].value_counts().index
```

```python
fact['gross_amount'] = fact['quantity'] * fact['price']
```

```python
np.random.seed(42)
fact['discount_amount'] = (fact['gross_amount'] * 
                               np.random.uniform(0, 0.2, len(fact))).round(2)

# Step 4: Derive remaining measures
fact['net_amount'] = fact['gross_amount'] - fact['discount_amount']

# Simulate cost (e.g. 60% of net as COGS)
fact['cost_amount'] = (fact['net_amount'] * np.random.uniform(0.5,0.8)).round(2)
fact['profit_amount'] = round(fact['net_amount'] - fact['cost_amount'], 2)
```
